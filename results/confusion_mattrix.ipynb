<span style="font-size: 24px;">Confusion Matrix</span>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Loading the data set
Train = np.load("../preprocessed_data/test_data.npz")
X_test = Train['X_test']
y_test = Train['y_test']
```

<span style="font-size: 24px;">Confusion Matrix using different cuts</span>


Note : {-1, 1} of actual data set normalized to {0, 1} during preprocessing

```python
# Prediction using different cuts are loaded and confusion mattrix is drawn to see which cut did well. The explanation of what cuts mean is given in the predict.py script,

cuts = [0.3, 0.4, 0.5, 0.6, 0.7]

for i in range(len(cuts)):
    predict = np.load(f"../predictions/predict_cut_{cuts[i]}.npz")
    y_pred = predict['pred']
    
    cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])

    disp.plot(cmap=plt.cm.Blues)
    plt.title(f"predict_cut_{cuts[i]}")
    plt.show()
```

A cut of 0.5 was chosen based one the confusion mattrix

```python
best_predict = np.load(f"../predictions/predict_cut_{0.5}.npz")
y_pred_best = best_predict['pred']
```

<span style="font-size: 24px;">Classification Report for cut 0.5</span>

```python
from sklearn.metrics import classification_report

# report classification 
print(classification_report(y_test, y_pred_best))
```
