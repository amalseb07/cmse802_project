<span style="font-size: 24px;">Testing the prediction of Regression</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import os
```

```python
# loading the train data set
train = np.load("../preprocessed_data_reg/train_data.npz")

X_train= train["X_train"]
y_train= train["y_train"]
```

```python
# loading the test data set
test = np.load("../preprocessed_data_reg/test_data.npz")

X_test= test["X_test"]
y_test= test["y_test"]
```

```python
# loading the weights for the best model and predict on the test and train set

model_path = "../src/best_model/best_cnn_model_reg.h5"
model = keras.models.load_model(model_path)

y_train_pred = model.predict(X_train, verbose=0) 
y_test_pred = model.predict(X_test, verbose=0) 
```

<span style="font-size: 24px;">Plot of Predicted vs True value</span>

- Predicted value of Train set plotted against true value 
- Predicted value of Test set plotted against true value 

```python


plt.scatter(y_train,y_train_pred, label='train set')
plt.scatter(y_test,y_test_pred,color='black',label='test set')

# line to compare the prediction
x=np.linspace(1,5,100)
y=x
plt.plot(x,y, color='red',label='x=y line')

plt.legend()
```
